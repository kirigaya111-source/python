```python
n = int(input("Nhập n: "))
tong = 0

for i in range(n):
    x = float(input(f"Nhập số thứ {i+1}: "))
    tong += x

trung_binh = tong / n
print("Trung bình cộng:", trung_binh)
```

```python
n = int(input("Nhập n: "))
dem = 0

for i in range(n):
    x = int(input(f"Nhập số thứ {i+1}: "))
    if x % 2 == 0:
        dem += 1

print("Số lượng số chẵn:", dem)
```

```python
n = int(input("Nhập n: "))

x = float(input("Nhập số thứ 1: "))
max_val = x
min_val = x

for i in range(2, n + 1):
    x = float(input(f"Nhập số thứ {i}: "))
    if x > max_val:
        max_val = x
    if x < min_val:
        min_val = x

print("Giá trị lớn nhất:", max_val)
print("Giá trị nhỏ nhất:", min_val)
```

```python
A = []

for i in range(5):
    x = float(input(f"Nhập số thứ {i+1}: "))
    A.append(x)

for i in range(4):
    for j in range(i + 1, 5):
        if A[i] > A[j]:
            A[i], A[j] = A[j], A[i]

print("Dãy số tăng dần:", A)
```

```python
def giaithua(n):
    if n == 0 or n == 1:
        return 1
    return n * giaithua(n - 1)

n = int(input("Nhập n: "))
print(f"{n}! =", giaithua(n))
```
